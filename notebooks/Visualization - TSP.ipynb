```python
from copy import deepcopy
from types import SimpleNamespace

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.reload_library()
plt.style.use('science')
from matplotlib import cm

from run import utils
```

```python
# init config
config = SimpleNamespace()
config.prob = "tsp"
config.nodes = 20
config.form = "dfj"
config.feat = 10
config.net = []
config.optm = "adam"
config.batch = 32
config.l1 = 0.0
config.l2 = 0.0
config.rel = False
config.lan = "gurobi"
config.proc = 32
config.path = "./res"
```

```python
# polynomial degree
degs = [1, 2, 4, 6]
```

```python
def getPath(config, mthd):
    config = deepcopy(config)
    if mthd == "lr":
        config.mthd = "2s"
        config.pred = "lr"
    if mthd == "rf":
        config.mthd = "2s"
        config.pred = "rf"
    if mthd == "spo":
        config.mthd = "spo"
        config.lr = 1e-2
    if mthd == "bb":
        config.mthd = "bb"
        config.lr = 5e-2
        config.smth = 20
    path = utils.getSavePath(config)
    return path
```

```python
def getDf(config, degs, mthd):
    dfs = pd.DataFrame()
    for deg in degs:
        config.deg = deg
        df = pd.read_csv(getPath(config, mthd))
        dfs[deg] = df["Unamb SPO"]
    return dfs
```

```python
def drawPlot(config, data, noise):
    # set config
    config = deepcopy(config)
    config.data = data
    config.noise = noise
    # color map
    cmap = cm.get_cmap("Set2")
    colors = cmap(np.linspace(0, 1, 8))
    # get df
    df_lr  = getDf(config, degs, "lr")
    df_rf  = getDf(config, degs, "rf")
    df_spo = getDf(config, degs, "spo")
    df_bb  = getDf(config, degs, "bb")
    # draw boxplot
    plt.figure(figsize=(16,6))
    c = colors[0]
    bp1 = plt.boxplot(df_lr, boxprops=dict(facecolor=c, color=c, linewidth=2), medianprops=dict(color="w", alpha=0.7),
                      whiskerprops=dict(color=c, linewidth=2), capprops=dict(color=c, linewidth=2),
                      flierprops=dict(markeredgecolor=c, marker="o", markersize=5, markeredgewidth=2),
                      patch_artist=True, positions=np.arange(df_spo.shape[1])-0.34, widths=0.18)
    c = colors[1]
    bp2 = plt.boxplot(df_rf, boxprops=dict(facecolor=c, color=c, linewidth=2), medianprops=dict(color="w", alpha=0.7),
                      whiskerprops=dict(color=c, linewidth=2), capprops=dict(color=c, linewidth=2),
                      flierprops=dict(markeredgecolor=c, marker="o", markersize=5, markeredgewidth=2),
                      patch_artist=True, positions=np.arange(df_spo.shape[1])-0.12, widths=0.18)
    c = colors[2]
    bp3 = plt.boxplot(df_spo, boxprops=dict(facecolor=c, color=c, linewidth=2), medianprops=dict(color="w", alpha=0.7),
                      whiskerprops=dict(color=c, linewidth=2), capprops=dict(color=c, linewidth=2), 
                      flierprops=dict(markeredgecolor=c, marker="o", markersize=5, markeredgewidth=2),
                      patch_artist=True, positions=np.arange(df_spo.shape[1])+0.12, widths=0.18)
    c = colors[3]
    bp4 = plt.boxplot(df_bb, boxprops=dict(facecolor=c, color=c, linewidth=2), medianprops=dict(color="w", alpha=0.7),
                      whiskerprops=dict(color=c, linewidth=2), capprops=dict(color=c, linewidth=2),
                      flierprops=dict(markeredgecolor=c, marker="o", markersize=5, markeredgewidth=2),
                      patch_artist=True, positions=np.arange(df_spo.shape[1])+0.34, widths=0.18)
    # vertical line
    plt.axvline(x=0.5, color="k", linestyle="--", linewidth=1.5)
    plt.axvline(x=1.5, color="k", linestyle="--", linewidth=1.5)
    plt.axvline(x=2.5, color="k", linestyle="--", linewidth=1.5)
    # labels and ticks
    plt.xlabel("Deg", fontsize=36)
    plt.xticks(ticks=[0,1,2,3], labels=[1,2,4,6], fontsize=28)
    plt.ylabel("Normalized Regret", fontsize=36)
    plt.yticks(fontsize=24)
    plt.xlim(-0.5, 3.5)
    plt.ylim(-0.02, 0.58)
    plt.title("Training Set Size = {},\nNoise Halfâˆ’width = {}".format(config.data, config.noise), fontsize=30)
    plt.legend([bp1["boxes"][0], bp2["boxes"][0], bp3["boxes"][0], bp4["boxes"][0]],
               ["2-stage LR", "2-stage RF", "SPO+","DBB"], fontsize=22, loc=2, labelspacing=0.2, handlelength=1)
    plt.show()
```

```python
drawPlot(config, data=100, noise=0.0)
```
