```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# number of cores
cores = [1,2,4,8,16,32]
```

```python
elapsed = []
for c in cores:
    df = pd.read_csv("./res/sp/h5w5/gurobi/procs/n1000p5-d4-e0.0_spo_lr_adam0.001_bs32_l10.0l20.0_c{}.csv".format(c))
    elapsed.append(df["Elapsed"].mean() / df["Epochs"].mean())
```

```python
plt.figure(figsize=(16,8))
plt.plot(cores, elapsed, marker='o', linestyle='dashed', markersize=8, color="g")
plt.xlabel("Num of Cores", fontsize=24)
plt.ylabel("Time Elpase per Epoch", fontsize=24)
plt.xticks(ticks=cores, fontsize=18)
plt.show()
```
